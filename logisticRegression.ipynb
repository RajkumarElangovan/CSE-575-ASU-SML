```python
#Method to extract features
def extractFeatures(df):
    X1=[]
    X2=[]
    for i in range(0,len(df)):
        X1.insert(i,np.mean(df[i]))
        X2.insert(i,np.std(df[i]))
    return X1,X2
```

```python
#Sigmoid aka Logistic Function
def sigmoid(W,X):
    z = np.dot(W.T,X)
    return 1/(1+np.exp(-z))
```

```python
#Function to update weights for each epoch
def weightCalc(X,Y,W,learnRate):
    Sig = sigmoid(W,X)
    for i in range(len(w)):
        gradient = X[i]*(Y-Sig)
        W[0][i] = W[0][i] + learnRate * gradient
    return W
```
